```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline

plt.scatter(df.area,df.price,color = "Red",marker='*')
plt.xlabel('Area(Sqr Ft)')
plt.ylabel('Price(US$)')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df['price'])
```

```python
reg.predict([[5500]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
!Y = M*x + C
135.78767123*5500+180616.43835616432
```

```python
%matplotlib inline
plt.scatter(df.area,df.price,color = "Red",marker='*')
plt.xlabel('Area',fontsize = 20)
plt.ylabel('Price',fontsize = 20)
plt.plot(df.area,reg.predict(df[['area']]),color = 'blue')
```

```python
d = pd.read_csv('areas.csv')
d.head()
```

```python
Prediction = reg.predict(d)
```

```python
d['prices']= Prediction
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.rename(columns={'per capita income (US$)': 'income_per_capita'},inplace=True)
df.head()
```

```python
plt.scatter(df.year,df.income_per_capita,color = 'Blue')
plt.xlabel("Year",fontsize=15)
plt.ylabel("Income Capita",fontsize=15)
```

```python
Regression = linear_model.LinearRegression()
Regression.fit(df[['year']],df['income_per_capita'])
```

```python
Regression.predict([[2020]])
```

```python
plt.scatter(df.year,df.income_per_capita,color = 'Blue')
plt.xlabel("Year",fontsize=15)
plt.ylabel("Income Capita",fontsize=15)
plt.plot(df.year,Regression.predict(df[['year']]),color = 'red')
```
